```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen import Lattice
from matplotlib import pyplot 
import numpy as np
from pprint import pprint
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.analysis.diffraction.core import AbstractDiffractionPatternCalculator as ADP
from IPython.display import Markdown,display,Image

with MPRester("BmFlV0IoLySTXxcf") as m:
    
    #structure for material id
    structure = m.get_structure_by_material_id("mp-672285")
    
    w = CifWriter(structure)
    w.write_file('mp-672285.cif')
    
    #cration de l'objet
    a = XRDCalculator()
    #calcul des pics de diffraction
    b = a.get_pattern(structure)
    a.get_plot(structure)
    pyplot.show()
    
```

```python
#calcul du premier pic
i1 = b.y[0] #intensit
angle1 = b.x[0] #2theta
hkl1 = b.hkls[0] #indices de Miller

#calcul du deuxime pic
i2 = b.y[1]
angle2 = b.x[1]
hkl2 = b.hkls[1]

#calcul du troisime pic
i3 = b.y[2]
angle3 = b.x[2]
hkl3 = b.hkls[2]
```

```python
print('premier pic du diffractogramme')
print('intensit :',i1) 
print('2\u03B8 :',angle1)
print('hkl :',hkl1)
```

```python
print('deuxime pic du diffractogramme')
print('intensit :',i2) 
print('2\u03B8 :',angle2)
print('hkl :',hkl2)
```

```python
print('premier pic du diffractogramme')
print('intensit :',i3) 
print('2\u03B8 :',angle3)
print('hkl :',hkl3)
```

```python
#diffractogramme des 3 premiers pics
xaxis = [angle1,angle2,angle3]
yaxis = [i1,i2,i3]
width = 0.25
pyplot.bar(xaxis,yaxis,width,color=(0.3,0.5,0.7,1.0))
pyplot.xlabel('2\u03B8')
pyplot.ylabel('Intensit')
pyplot.title('Diffractogramme des trois premiers pics')
pyplot.show()
```
